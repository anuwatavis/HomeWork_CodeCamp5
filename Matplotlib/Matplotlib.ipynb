### Quiz 5

```python
x = np.arange(1, 20)
y1 = x**2
y2 = x**3
plt.plot(x, y1, c='r', marker='o')
plt.plot(x, y2, c='b', marker='v')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Exponential Plotting')
plt.show()
```

### Quiz 6

```python
x = np.arange(1, 20)
y1 = x**2 
y2 = x**(1/2)
plt.plot(x, y1, c='r', marker='o', label='y = x^2')
plt.plot(x, y2, c='b', marker='v', label='y = x^1/2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.title('Exponential Plotting')
plt.show()
```

### Quiz 7

```python
y = np.arange(-15, 15)
x = y**2 + 4*y 
plt.plot(x, y, c='#ffff80', marker='o', label='x = y^2 + 4*y')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim([-5, 150])
plt.grid()
plt.legend()
plt.title('X = Y^2 + 4*Y')
plt.show()
```

### Quiz 8

```python
y = np.arange(0, 20)
x = y**3

x1 = np.arange(0, 20)
y1 = x1**2 + 2

fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.1, 1, 1])
ax1.plot(x, y, label= 'x = y ^3')
ax1.set_title('X = Y ^ 3')
ax1.set(xlabel = 'X', ylabel ='Y')
ax1.legend(loc=0)

ax2 = fig.add_axes([0.6, 0.35, 0.3, 0.3])
ax2.plot(x1, y1, label= 'y = x^2 +2')
ax2.set_title('Y = X^2 + 2')
ax2.set(xlabel = 'X', ylabel ='Y')
ax2.legend(loc=0)

plt.show()
```

### Quiz 9

```python
#importfile titanic 
df = pd.read_csv('train.csv')
df
```

```python
pclass_count = df['Pclass'].value_counts().sort_index().sort_index()
pclass_count
```

```python
pclass1_ratio = pclass_count[1] / pclass_count.sum() 
print("Pclass 1 มีอัตราส่วนเท่ากับ",pclass1_ratio )
pclass2_ratio = pclass_count[2] / pclass_count.sum() 
print("Pclass 2 มีอัตราส่วนเท่ากับ",pclass2_ratio )
pclass2_ratio = pclass_count[3] / pclass_count.sum() 
print("Pclass 3 มีอัตราส่วนเท่ากับ",pclass2_ratio )
```

```python
lst_pclass_ratio = list(pclass_count)/pclass_count.sum()
lst_pclass_ratio
```

```python
plt.figure(figsize = [5, 5])
labels = ['Pclass1', 'PClass2', 'PClass3']
plt.pie(lst_pclass_ratio, labels = labels)
plt.show()
```

### Quiz 15

```python
name = df['Name']
lst_lastname = []
for i in name:
    lst_lastname.append(i.split(',')[0])
lst_lastname
lastname_series = pd.Series(lst_lastname)
count_duplicate_lastname = lastname_series.value_counts()
```

```python
result = count_duplicate_lastname.groupby(count_duplicate_lastname).count()
```

```python
duplicate_index = result[result.index > 1]
lst_index = list(duplicate_index.index)
lst_data = list(result.loc[lst_index])
plt.bar(lst_index, lst_data, color='salmon')
plt.show()
```
