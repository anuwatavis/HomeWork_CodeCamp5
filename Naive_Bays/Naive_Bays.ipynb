```
def displot_plot(data_frame, row, column):
  f, axes = plt.subplots(row, column, figsize=(12, 16))
  count = 0 
  for i in range(row) : 
    for j in range(column):
      sns.distplot(data_frame.iloc[:, count], ax=axes[i,j])
      count = count + 1
(displot_plot(df, 5,2 ))
```

8. สร้าง scatter plot ของความสัมพันธ์ที่มี Correlation สูงสุด

```
def find_max_corr(dataframe) :
  dataframe_corr = dataframe.corr().abs()
  sorted_correlation = dataframe_corr.unstack().sort_values(ascending = False)
  return sorted_correlation
sort_correlation = find_max_corr(df)
sort_correlation
```

```
labels = df['Type']
labels
```

```
#Default
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
X_train, X_test, y_train, y_test = train_test_split(df.drop(['Type'], axis =1), labels, test_size = 0.3, random_state = 1)
```

```
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```
#Default parameters of naive bays
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train, y_train)
predicted = nb.predict(X_test)
confusion_matrix(y_test, predicted)
```

```
print('Accuracy : ', accuracy_score(y_test, predicted))
print('F1 score : ', f1_score(y_test, predicted, average='micro'))
print('Precision : ', precision_score(y_test, predicted, average='micro'))
print('Recall :', recall_score(y_test, predicted, average='micro') )
```

```
#Normalization
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()
df_min_max = min_max_scaler.fit_transform(df.drop(['Type'], axis=1))
df_min_max
```

```
norm_df  = pd.DataFrame(df_min_max, columns=df.columns[:-1])
```

```
norm_df
```

```
X_train, X_test, y_train, y_test = train_test_split(norm_df, labels, test_size = 0.3, random_state = 1)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```
nb_2 = GaussianNB()
nb_2.fit(X_train, y_train)
predicted_2 = nb_2.predict(X_test)
predicted_2
confusion_matrix(y_test, predicted_2)
```

```
print('Accuracy : ', accuracy_score(y_test, predicted_2))
print('F1 score : ', f1_score(y_test, predicted_2, average='micro'))
print('Precision : ', precision_score(y_test, predicted_2, average='micro'))
print('Recall :', recall_score(y_test, predicted_2, average='micro') )

result = [[ f1_score(y_test, predicted_2, average='micro')]]
summary = pd.DataFrame(result,columns=['F1 score'],index=['Normalization (All Feature) Before Split'])
summary
```

```
#Standardization
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df_scale = sc.fit_transform(df.drop(['Type'], axis =1))
sc_df = pd.DataFrame(df_scale, columns = df.columns[:-1])
sc_df
```

```
X_train, X_test, y_train, y_test = train_test_split(sc_df, labels, test_size = 0.3, random_state = 1)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```
nb_3 = GaussianNB()
nb_3.fit(X_train, y_train)
predicted_3 = nb_3.predict(X_test)
predicted_3
confusion_matrix(y_test, predicted_3)
```

```
print('Accuracy : ', accuracy_score(y_test, predicted_3))
print('F1 score : ', f1_score(y_test, predicted_3, average='micro'))
print('Precision : ', precision_score(y_test, predicted_3, average='micro'))
print('Recall :', recall_score(y_test, predicted_3, average='micro') )
result = [[ f1_score(y_test, predicted_3, average='micro')]]
prepare = pd.DataFrame(result,columns=['F1 score'],index=['Standardization (All Feature) Before Split'])
summary = summary.append(prepare)
summary
```
