```python
def find_max_corr(dataframe) :
  dataframe_corr = dataframe.corr().abs()
  sorted_correlation = dataframe_corr.unstack().sort_values(ascending = False)
  return sorted_correlation

sort_correlation = find_max_corr(cancer_data)
sort_correlation
# result of max correlation value between feature are sqft_living and sqft_above
# result of min correlation value between feature are waterfront and condition
```

```python
target = [1 if x =='M' else 0 for x in cancer_data['diagnosis_result']]
diagnosis_result = pd.DataFrame(target, columns=['result'])
diagnosis_result
```

```python
#feature selection
cancer_data.columns
```

```python
X = cancer_data[['radius', 'texture', 'perimeter', 'area','smoothness', 'compactness', 'symmetry', 'fractal_dimension']]
y = diagnosis_result
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
sns.distplot(X_train['area'])
```

```python
sns.distplot(X_test['area'])
```

```python
X
```

```python
#Normalization
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max_scaler = MinMaxScaler()
```

```python
cancer_data_normalization = min_max_scaler.fit_transform(X)
cancer_data_normalization
```

```python
df_cancer_data_normalization =   pd.DataFrame(cancer_data_normalization, columns=X.columns)
df_cancer_data_normalization 
```

# Training Model

```python
from sklearn.svm import SVC
```

```python
svc = SVC()
```

```python
svc.fit(X_train, y_train)
```

```python
predicted = svc.predict(X_test)
```

```python
predicted
```

## Model Evaluation

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
```

```python
confusion_matrix(y_test, predicted)
```

Thus in binary classification, the count of   
true negatives is :math:`C_{0,0}`, false negatives is :math:`C_{1,0}`, true positives is    
:math:`C_{1,1}` and false positives is :math:`C_{0,1}`.  

```python
#Normalization 
print('Accuracy = ', accuracy_score(y_test, predicted))
print('F1 Score = ', f1_score(y_test, predicted))
print('Preciscion = ', precision_score(y_test, predicted))
print('Recall = ', recall_score(y_test, predicted))
result = [[accuracy_score(y_test,predicted),f1_score(y_test,predicted),precision_score(y_test,predicted),recall_score(y_test,predicted)]]
summary = pd.DataFrame(result,columns=['Accuracy','F1 score','Recall','Precision'],index=['Normalization (All Feature)'])
summary
```

```python
#hyperparameter tuning
```

```python
from sklearn.model_selection import  GridSearchCV
```

```python
param_combination = {'C' : [0.01, 0.1, 1, 10, 100, 100, 1000, 10000], 'gamma': [0.00001, 0.0001, 0.001, 0.01, 0.1, 1, 10] }
```

```python
grid_search = GridSearchCV(SVC(), param_combination, verbose=1)
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_estimator_
```

```python
grid_prediceted = grid_search.predict(X_test)
grid_prediceted
```

```python
confusion_matrix(y_test, grid_prediceted)
```

```python
print('Accuracy = ', accuracy_score(y_test, grid_prediceted))
print('F1 Score = ', f1_score(y_test, grid_prediceted))
print('Preciscion = ', precision_score(y_test, grid_prediceted))
print('Recall = ', recall_score(y_test, grid_prediceted))
result = [[accuracy_score(y_test,grid_prediceted),f1_score(y_test,grid_prediceted),precision_score(y_test,grid_prediceted),recall_score(y_test,grid_prediceted)]]
prepare = pd.DataFrame(result,columns=['Accuracy','F1 score','Recall','Precision'],index=['Hyperparameter Tuning (All Feature)'])
summary = summary.append(prepare)
summary
```
