```

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
```

12. สร้าง train/test split ของบ้าน สามารถลองทดสอบ 70:30, 80:20, 90:10 ratio ได้ตามใจชอบ

13. ทำ Data Transformation  Data Scaling


```

X = np.array(df['sqft_living']).reshape(-1, 1)
y = np.array(df['price']).reshape(-1, 1)
# data scaling 
sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y)
```

```
#train/test (80:20)
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,random_state = 100)
```

14. เทรนโมเดลแบบ Linear และ rbf ของคู่ที่ผู้เรียนคิดว่าเหมาะสม
(ให้เป็น 1 dependent VS 1 independent)


```
#kenel = rbf
#independent variable : sqft_living
#dependent variable : price
linear_regressor = SVR(kernel='linear')
linear_regressor.fit(X_train, y_train)
```

```
#kenel = rbf
#independent variable : sqft_living
#dependent variable : price
rbf_regressor = SVR(kernel='rbf')
rbf_regressor.fit(X_train, y_train)
```

15. ทดสอบโมเดลวัดค่า MAE, MSE, RMSE

```
result_linear = linear_regressor.predict(X_test)
#MAE, MSE, RMSE (Linear Model)
MEA_linear = metrics.mean_absolute_error(y_test,result_linear)
MSE_linear = metrics.mean_squared_error(y_test, result_linear)
RMSE_linear = np.sqrt(metrics.mean_squared_error(y_test, result_linear))
print('MEA = ', MEA_linear)
print('MSE = ', MSE_linear)
print('RMSE = ', RMSE_linear)
compare_error = []
compare_error.append(['1 dependent : kernel = linear',MEA_linear, MSE_linear, RMSE_linear])
```

```
result_rbf = rbf_regressor.predict(X_test)
#MAE, MSE, RMSE (Rbf Model)
MEA_rbf = metrics.mean_absolute_error(y_test,result_rbf)
MSE_rbf = metrics.mean_squared_error(y_test, result_rbf)
RMSE_rbf = np.sqrt(metrics.mean_squared_error(y_test, result_rbf))
print('MEA = ', MEA_rbf)
print('MSE = ', MSE_rbf)
print('RMSE = ', RMSE_rbf)
compare_error.append(['1 dependent : kernel = rbf', MEA_rbf, MSE_rbf, RMSE_rbf])
```

17. เทรนโมเดลแบบ Simple Linear Regression และ Multiple Linear Regression
(สามารถเลือกคู่ได้ หรือ จะลองทุกคู่ก็ได้ ตามความคิดของผู้เรียน)

18. ทดสอบโมเดลวัดค่า MAE, MSE, RMSE ของ Linear Regression

**Simple Linear Regression **


```
#data transformation
X = np.array(df['sqft_living']).reshape(-1, 1)
y = np.array(df['price']).reshape(-1, 1)

#data scaling
X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y)

X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,random_state = 100)

#model traing
simple_linear_model = LinearRegression()
simple_linear_model.fit(X_train, y_train)
print('intercept and coef ')
simple_linear_model.intercept_ , simple_linear_model.coef_
```

```
simple_linear_result = simple_linear_model.predict(X_test)
#MAE, MSE, RMSE (Rbf Model)
MEA = metrics.mean_absolute_error(y_test,simple_linear_result)
MSE = metrics.mean_squared_error(y_test, simple_linear_result)
RMSE = np.sqrt(metrics.mean_squared_error(y_test, simple_linear_result))
print('MEA = ', MEA)
print('MSE = ', MSE)
print('RMSE = ', RMSE)
compare_error.append(['linear regression', MEA, MSE, RMSE])
```

*Multiple Linear Regression*

```
#data transformation
X = df[['bedrooms', 'sqft_living','sqft_above','bathrooms', 'sqft_basement' ]]
y = np.array(df['price']).reshape(-1, 1)

#data scaling
X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y)

X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,random_state = 100)

#model traing
multiple_linear_model = LinearRegression()
multiple_linear_model.fit(X_train, y_train)
print('intercept and coef ')
multiple_linear_model.intercept_ , multiple_linear_model.coef_
```

```
multiple_linear_result = multiple_linear_model.predict(X_test)
#MAE, MSE, RMSE (Rbf Model)
MEA = metrics.mean_absolute_error(y_test,multiple_linear_result)
MSE = metrics.mean_squared_error(y_test, multiple_linear_result)
RMSE = np.sqrt(metrics.mean_squared_error(y_test, multiple_linear_result))
print('MEA = ', MEA)
print('MSE = ', MSE)
print('RMSE = ', RMSE)
compare_error.append(['multi linear regression', MEA, MSE, RMSE])
```
