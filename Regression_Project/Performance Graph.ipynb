# 84.  สร้าง bar chart แสดงค่า RMSE ของทุกผลลัพธ์ทั้งหมดที่ทำมา ยกเว้น 1 indepedent variable (สองอันแรก) แล้วเปรียบเทียบว่า Model มีประสิทธิภาพมากที่สุด


```
X = ['Linear Regression', 'Support Vector Regression', 'Decision Tree', 'Random Forest']
y = [1069, 1124, 1472, 1095]
```

```
fig = plt.figure(figsize=(12, 8))
sns.barplot(x = X, y = y)
plt.title('RMSE of four modelsm (less is better)')
plt.xlabel('Models Titles')
plt.ylabel('RMSE score')
```
