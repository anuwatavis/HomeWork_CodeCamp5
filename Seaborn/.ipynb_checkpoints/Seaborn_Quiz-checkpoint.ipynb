```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
### Quiz 1
```

```python
index = ['A', 'B', 'C', 'D']
columns = ['2014', '2015', '2016', '2017']
values = np.random.random_sample((4,4))*500e6 + 50e6
sales_df = pd.DataFrame(values, index, columns)
sales_df
```

```python
### Quiz 2
```

```python
mean = [sales_df.loc['A'].mean(), sales_df.loc['B'].mean(), sales_df.loc['C'].mean(), sales_df.loc['D'].mean()]
```

```python
fig = plt.figure(figsize=[5, 5])
sns.barplot(x = sales_df.index, y = mean )
```

```python
### Quiz 4 
```

```python
fig = plt.figure(figsize=[5, 5])
sns.scatterplot(x = sales_df.index, y = '2017', data = sales_df)
```

```python
fig = plt.figure(figsize=[5, 5])
sns.scatterplot(x = '2017', y = sales_df.index, data = sales_df)
```
