## 16. วัดผลโมเดล โดยใช้ confusion matrix และ ประเมินผลด้วยคะแนน Accuracy, F1 score, Recall, Precision


```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
```

# Use Standardization_data

```python
#Use Normalization_Data
from sklearn.model_selection import train_test_split
X_train , X_test,y_train,y_test = train_test_split(standardization_data,lable_data,train_size = 0.7,random_state=32)
KNN = KNeighborsClassifier()
KNN.fit(X_train, y_train)
```

```python
predicted = KNN.predict(X_test)
```

```python
confusion_matrix(y_test,predicted )
```

```python
print("Accuracy", accuracy_score(y_test, predicted))
print("F1_Score", f1_score(y_test, predicted, average = 'micro'))
print("Precision", precision_score(y_test, predicted, average = 'micro'))
print("Recall", recall_score(y_test, predicted, average = 'micro'))
```

## 17. หาค่า K ที่ดีที่สุด สำหรับ Dataset นี้


```python
accuracy_lst = []
for i in range (1, 30):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    predicted_i = knn.predict(X_test)
    accuracy_lst.append(accuracy_score(y_test, predicted_i))
```

```python
fig = plt.figure(figsize = (10, 8))
plt.plot(range(1,30), accuracy_lst, color='black', linestyle='dashed', marker = 'o', markerfacecolor = 'blue')
plt.xlabel("Number of K")
plt.ylabel("Accurcy")
```
